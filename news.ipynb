#### 기사 내용가져오기 합본

```python
from bs4 import BeautifulSoup
import requests
import time
import bardapi
import os
import re

def fetch_webpage(url):
    response = requests.get(url)
    if response.status_code == 200:
        return response.content
    else:
        return None


def parse_content(html_content):
    soup = BeautifulSoup(html_content, 'html.parser')
    # 예시로 모든 <p> 태그의 텍스트를 추출합니다.
    base_selector = "#container > section > div > section > div.sec_body > div > ul > li:nth-child({}) > a"
    hrefs = []
    for i in range(1, 11):  # 1부터 10까지
        selector = base_selector.format(i)  # 각 li에 대한 선택자 생성
        a_tag = soup.select_one(selector)
        if a_tag:
            hrefs.append(a_tag['href'])
    print(hrefs)
    return hrefs

# news.ipynb에서 가져온 코드 모듈화


def fetch_article_details(url_list):
    title = []
    datetime = []
    content = []
    summary = []
    bard_api_key = 'bQgJ7MlGfJj0t-bujqVvAAN-csVt_sXuvtNRxIkrBEO1ilcG4hyyEeaW6nUQf9gWkFoZKQ.'
    os.environ['_BARD_API_KEY'] = bard_api_key
    for url in url_list:
        response = requests.get(url)
        soup = BeautifulSoup(response.content, 'html.parser')
        if soup.find('div', class_='locked_wrapper'):
            continue
        title_ClassName = "news_ttl"
        individual_title = soup.find(
            'h2', class_=title_ClassName).get_text().strip()
        individual_datetime = soup.find(
            'dl', class_='registration').find('dd').get_text().strip()
        individual_content = soup.find(
            'div', class_='news_cnt_detail_wrap')
        individual_content_list = []
        for i in individual_content:
            text = i.get_text().replace('“', '').replace('”', '').replace('  ','').replace('\n','')
            individual_content_list.append(text)
        individual_contents = ' '.join(individual_content_list)
        individual_contents = re.sub(r"[^가-힣0-9A-Za-z\s]", "", individual_contents)
        template = f"""150자 이내로 요약 {individual_contents}"""
        response = bardapi.core.Bard(timeout=10).get_answer(template)
        individual_summary = response['content']
        summary.append(individual_summary)
        title.append(individual_title)
        datetime.append(individual_datetime)
        content.append(individual_contents)
        time.sleep(2)
    data = {'title': title, 'datetime': datetime, 'content': content,'summary': summary}
    return data


url = 'https://www.mk.co.kr/news/ranking/economy/'  # 여기에 원하는 웹페이지 URL을 넣어주세요
html_content = fetch_webpage(url)
if html_content:
    url_list = parse_content(html_content)
    daily_economy_news = fetch_article_details(url_list)
    print(daily_economy_news)
    print(daily_economy_news['title'])
else:
    print("Failed to retrieve the webpage.")
```

```python
import requests
from bardapi.constants import SESSION_HEADERS
from bardapi import Bard
import json
import os

def load_cookie(element_name):
    with open(os.getcwd() + "/" + "bard.json", "r", encoding="utf-8") as file:
        data = json.load(file)
        if isinstance(data, list):
            for item in data:
                if item.get("name") == element_name:
                    return item.get("value")
        return None

_1PSID = load_cookie("__Secure-1PSID")
_1PSIDTS = load_cookie("__Secure-1PSIDTS")
_1PSIDCC = load_cookie("__Secure-1PSIDCC")
session = requests.Session()
session.headers = SESSION_HEADERS
session.cookies.set("__Secure-1PSID", _1PSID)
session.cookies.set("__Secure-1PSIDTS", _1PSIDTS)
session.cookies.set("__Secure-1PSIDCC", _1PSIDCC)

bard = Bard(token=_1PSID, session=session)
print(bard.get_answer("gnomes")['content'])
```

```python
from langchain.chat_models import ChatOpenAI
from langchain.llms import OpenAI
from typing import Optional, List, Dict
from dotenv import load_dotenv  # .env file에서 환경변수를 불러옴
from bs4 import BeautifulSoup
import requests
import time
load_dotenv()
chat_model = ChatOpenAI(model_name="gpt-3.5-turbo")

def fetch_webpage(url: str) -> Optional[bytes]: 
    response = requests.get(url)
    if response.status_code == 200:
        return response.content
    else:
        return None


def parse_content(html_content: bytes) -> List[str]:
    soup = BeautifulSoup(html_content, 'html.parser')
    # 예시로 모든 <p> 태그의 텍스트를 추출합니다.
    base_selector = "#container > section > div > section > div.sec_body > div > ul > li:nth-child({}) > a"
    hrefs = []
    for i in range(1, 11):  # 1부터 10까지
        selector = base_selector.format(i)  # 각 li에 대한 선택자 생성
        a_tag = soup.select_one(selector)
        if a_tag:
            hrefs.append(a_tag['href'])
    print(hrefs)
    return hrefs

# OpenAI API를 사용하여 요약
def summarize_ai(to_summarize: str) -> str:
    result = chat_model.predict(f"""
                            이제부터 너는 기사를 요약해주는 봇이야. 대답할 필요없이
                            아래의 기사를 읽고, 200자 이내로 요약을 부탁할게.
                            {to_summarize}
                            형식은 다음과 같아!
                            요약부분(200자이내)
                            """)
    return result

# news.ipynb에서 가져온 코드 모듈화
def fetch_article_details(url_list: list) -> dict:
    title = []
    datetime = []
    content = []
    summary = []
    for url in url_list:
        response = requests.get(url)
        soup = BeautifulSoup(response.content, 'html.parser')
        if soup.find('div', class_='locked_wrapper'):
            continue
        title_ClassName = "news_ttl"
        individual_title = soup.find(
            'h2', class_=title_ClassName).get_text().strip()
        individual_datetime = soup.find(
            'dl', class_='registration').find('dd').get_text().strip()
        individual_content = soup.find(
            'div', class_='news_cnt_detail_wrap')
        individual_content_list = []
        for i in individual_content:
            text = i.get_text().replace('“', '').replace(
                '”', '').replace('  ', '').replace('\n', '')
            individual_content_list.append(text)
        individual_contents = ' '.join(individual_content_list)
        indivdual_summary = summarize_ai(individual_contents)
        title.append(individual_title)
        datetime.append(individual_datetime)
        content.append(individual_contents)
        summary.append(indivdual_summary)
        print(indivdual_summary)
        time.sleep(0.5)
    data = {'title': title, 'datetime': datetime,
            'content': content, 'summary': summary}
    return data


url = 'https://www.mk.co.kr/news/ranking/economy/'  # 여기에 원하는 웹페이지 URL을 넣어주세요
html_content = fetch_webpage(url)
if html_content:
    url_list = parse_content(html_content)
    daily_economy_news = fetch_article_details(url_list)
    print(daily_economy_news)
else:
    print("Failed to retrieve the webpage.")


```
